## Data Cleaning

```python
#Total Duplicates
df.duplicated().sum()
```

```python
#Rows after removing the duplicates
df.drop_duplicates(inplace=True)
unique_rows = df.shape[0]
print(f'Unique rows after removing duplicates: {unique_rows}')
```

```python
#null values
null_values = df.isnull().sum().sort_values(ascending=False)
print('Missing values summary:\n')
print(null_values)
```

```python
#Null values in percentage
missing_percentage = (null_values / unique_rows * 100)
print('\nMissing values percentage:\n')
print(missing_percentage)
```

```python
#fill null values in comorbidity column with mode
df['comorbidity'] = df['comorbidity'].fillna(df['comorbidity'].mode()[0])
```

```python
#fill null values in alcohol_use column with mode
df['alcohol_use'] = df['alcohol_use'].fillna(df['alcohol_use'].mode()[0])
```

```python
# Mean recovery days for recovered/alive patients
mean_recovery = df['recovery_days'].mean()

# Fill missing recovery_days based on death_flag
df.loc[(df['recovery_days'].isna()) & (df['death_flag'] == 0), 'recovery_days'] = mean_recovery

df.loc[(df['recovery_days'].isna()) & (df['death_flag'] == 1), 'recovery_days'] = 0
```

```python
#Cause of Death was removed,because it contained a very high percentage of missing values.
df.drop(columns=['cause_of_death'], inplace=True)
```

```python
#treatment_cost_inr data type changed to int
df['treatment_cost_inr'] = df['treatment_cost_inr'].astype(int)
```

```python
#Null values after cleaning
df.isnull().sum()
```

## Feature Engineering

```python
#New age_group column added
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 50, 65, 100],
                                   labels=['Child', 'Young Adult', 'Adult', 'Senior', 'Elderly'])
```

```python
df['patient_risk_level'] = df.apply(
    lambda row: 'High' if (row['death_flag'] == 1) else 'Low',
    axis=1
)
```

### Observations
#### State-wise Patient Distribution
- The highest number of patients in the dataset are from Uttar Pradesh.
- The lowest number of patients are from Chandigarh.
#### Average Treatment Cost Analysis

- Lung Cancer has the highest average treatment cost among all diseases, indicating a significant financial burden on patients.
- Diarrhea has the lowest average treatment cost, suggesting comparatively lower treatment expenses.
#### Alcohol Consumption Pattern
- The majority of patients reported occasional alcohol consumption, making it the most common alcohol usage category in the dataset.
#### Blood Group Distribution
- The most frequently occurring blood groups among patients are B+ and O+.
- These blood groups constitute a significant proportion of the patient population.
#### Income Level Analysis
- Most patients belong to the lower-income group.
- This suggests that a large portion of the recorded disease burden is associated with economically disadvantaged populations.
#### Smoking Status and Disease Severity
- Patients who reported never smoking constitute the largest group among those with severe disease conditions.
- This suggests that factors other than smoking may also play a significant role in disease severity within the dataset.
#### State-wise Mortality Analysis
- Uttar Pradesh records the highest number of patient deaths among all states in the dataset.
- This may be influenced by the larger number of patients represented from the state.
